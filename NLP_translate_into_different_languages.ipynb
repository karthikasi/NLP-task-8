```python
tamil=[]
for i in df3["English words/sentences"]:
  l=str(i)
  k=ts.google(l,from_language='en',to_language="ta")
  tamil.append("".join(k))

print(tamil)

#data=df3.to_dict()
#print(data)
```

```python
tam=pd.Series(tamil)
```

```python
tamizh=pd.DataFrame(tam,columns=["Tamil words"])
tamizh.head()
```

```python
spanish=[]
for i in df3["English words/sentences"]:
  l=str(i)
  k=ts.google(l,from_language='en',to_language='es')
  spanish.append("".join(k))
print(spanish)
```

```python
spain=pd.Series(spanish)
print(spain)
```

```python
spain=pd.DataFrame(spain,columns=["Sapnish words"])
spain.head()
```

```python
german=[]
for i in df3["English words/sentences"]:
  l=str(i)
  k=ts.google(l,from_language="en",to_language="de")
  german.append("".join(k))
print(german)
germanian=pd.Series(german)
```

```python
print(germanian)
```

```python
germ=pd.DataFrame(germanian,columns=["german words"])
germ.head()
```

```python
french=[]
for i in df3["English words/sentences"]:
  l=str(i)
  k=ts.google(l,from_languages="en",to_languages="fr")
  french.append("".join(k))

france=pd.Series(french)
france.head()
```

```python
fre=pd.DataFrame(france, columns=["French words"])
```
